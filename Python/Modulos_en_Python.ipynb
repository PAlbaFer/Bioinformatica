# MODULOS EN PYTHON

#### Un módulo en Python es simplemente un archivo .py que contiene funciones, variables o clases que puedes reutilizar en otros programas. Tienes un archivo llamado mimodulo.py

```python
import mimodulo

from importlib import reload # esto es para que si modificamos el módulo, nos lo actualice aquí. Sino hacemos esto, no lo hace.
reload(mimodulo)

print(mimodulo.saludar("Octavio"))      
print(mimodulo.pi)                  

persona = mimodulo.Persona("Rigoberta")
print(persona.presentar())       
```

```python
# Podemos crear una carpeta con varios módulos usando %%writefile para cada archivo:
# Aquí escribimos los módulos dentro del mismo jupyter
# Crear carpeta llamada mi_paquete (solo si no existe)
import os
os.makedirs("mi_paquete", exist_ok=True)

# __init__.py
with open("mi_paquete/__init__.py", "w") as f:
    f.write("from .saludos import hola, adios\nfrom .matematicas import sumar, multiplicar\n")

# saludos.py
with open("mi_paquete/saludos.py", "w") as f:
    f.write("""
def hola(nombre):
    return f"¡Hola, {nombre}!"

def adios(nombre):
    return f"Adiós, {nombre}!"
""")

# matematicas.py
with open("mi_paquete/matematicas.py", "w") as f:
    f.write("""
def sumar(a, b):
    return a + b

def multiplicar(a, b):
    return a * b
""")
```
