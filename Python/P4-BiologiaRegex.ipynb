##### 4.- Escribe una función que cuente el porcentaje de A de una secuencia

```python
dna_file.seek(0)
dna=dna_file.read()
def percent_A(dna): 
    if not dna:
        return 0
    total=len(dna)
    count_A=dna.upper().count('A')
    percent=(count_A/total)*100
    return percent
print(f'A: {percent_A(dna)}%')
```

##### ¿Puedes hacer que devuelva dos decimales? ¿Y que lidie con mayúsculas y minúsculas?

```python
print(f'A: {percent_A(dna):.2f}%')
```

##### ¿Y que devuelva el número de decimales que yo quiero?

```python
def percent_A(dna, decimal=2): 
    if not dna:
        return 0
    total=len(dna)
    count_A=dna.upper().count('A')
    percent=(count_A/total)*100
    return round(percent,decimal)
print(percent_A(dna,5))
```

##### ¿Sabes poner un valor por defecto en un argumento?

```python
print(f'A: {percent_A(dna):.2f}%')
```

##### ¿Qué pasa si no ponemos un argumento? ¿Sabrías hacer una función encapsulada?

```python
def percent_A(dna, decimal=2):
    def count_A(s):
        return s.upper().count('A')
    if not dna: 
        return 0 
    total=len(dna)
    percent=(count_A(dna)/total)*100
    return round(percent,decimal)
print(f'A: {percent_A(dna, 2)}%')
```
