#### 7.- Con la secuencia y la lista siguiente, escribe un algoritmo que cuente el número de veces que aparece un elemento de la lista en la secuencia. Pista: hacer un diccionario para almacenar los resultados puede ser buena idea

```python
dna = "AATGATGAACGAC" 
dinucleotides = ['AA','AT','AG','AC', 
                 'TA','TT','TG','TC', 
                 'GA','GT','GG','GC', 
                 'CA','CT','CG','CT'] 
```

```python
# Diccionario: 
counts = {dn: 0 for dn in dinucleotides}
# Contar dinucleotidos (recorrer la secuencia de 2 en 2):
for n in range(len(dna) - 1):
    dn = dna[n:n + 2]
    if dn in counts:
        counts[dn] += 1 
for dn, counts in counts.items():
    print(dn,counts)
```

##### ¿Cuántos counts tiene TC?

```python
counts = {dn: 0 for dn in dinucleotides}
for n in range(len(dna) - 1):
    dn = dna[n:n + 2]
    if dn in counts:
        counts[dn] += 1 
print("TC:",counts["TC"])
```

##### ¿Puedes imprimir los dinuleótidos que tengan counts igual a 2?

```python
counts = {dn: 0 for dn in dinucleotides}
for n in range(len(dna) - 1):
    dn = dna[n:n + 2]
    if dn in counts:
        counts[dn] += 1 
print("Dinucleótidos con counts = 2:")
for dn, counts in counts.items():
    if counts == 2:
        print(dn,counts)
```
